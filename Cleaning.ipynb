```python
import pandas as pd

file_to_load= "standard decks.csv"

standard_df=pd.read_csv(file_to_load)

standard_df
```

```python
Cleaned_df = standard_df["Deck list"].str.split(pat="\r\n", expand=True)
Cleaned_df
```

```python
Cleaned_df.dropna
```

```python
for each_col in Cleaned_df.columns: 
    Cleaned_df[each_col]=Cleaned_df[each_col].replace('\d +', '', regex=True)
Cleaned_df
```

```python
card_list=[]
for row_idx, row in Cleaned_df.iterrows(): 
    for col_idx, col in row.iteritems(): 
        if not col in card_list: 
            card_list.append(col)
            
```

```python
len(card_list)
```

```python
display('Sideboard' in card_list)
display(None in card_list)
```

```python
card_list.remove(None)
card_list.remove('Sideboard')
card_list.remove('')
```

```python
converted_df=pd.DataFrame(columns=card_list)
for row_idx, row in Cleaned_df.iterrows(): 
    deck=Cleaned_df.loc[row_idx].tolist()
    for each_card in deck: 
        if each_card: 
            converted_df.loc[row_idx, each_card]=1
```

```python
converted_df.head(20)
```

```python
converted_df=converted_df.fillna(0)
converted_df
```

```python
converted_df.sum()[converted_df.sum()==0]
```

```python
from sklearn.cluster import KMeans

inertia=[]

for i in range(1, 20): 
    km=KMeans(n_clusters=i)
    km.fit(converted_df)
    inertia.append(km.inertia_)
```

```python
km=KMeans(n_clusters=5)
km.fit(converted_df)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(2, 21), inertia)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=4)
pca.fit(converted_df)
pca_df=pca.transform(converted_df)
```

```python
pca_df
```

```python
fig=plt.figure(figsize=(10, 10))
ax=fig.add_subplot(projection='3d')
ax.scatter(pca_df[:, 0], pca_df[:, 1], pca_df[:, 2], c=km.labels_)
```

```python
Cleaned_df.to_csv("cleaned.csv")
```
